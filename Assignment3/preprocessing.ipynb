```python
# Fetch the data from CSV and load it to the DataFrame
import pandas as pd
df = pd.read_csv('Data.csv')
print(df)
```

```python
# Analyzing the data using Histogram

import matplotlib.pyplot as plt
df.hist()
plt.show()
```

```python
# fill the missing values using Simple Imputation using Pandas only.

df["Updated_Country"] = df["Country"]
df["Updated_Country"].fillna("No Country", inplace = True)
df["Updated_Age"] =  df["Age"]
df["Updated_Age"].fillna(df["Age"].mean(), inplace = True)
df["Updated_Salary"] = df["Salary"]
df["Updated_Salary"].fillna(df["Salary"].median(), inplace = True)
df["Updated_Purchased"] = df["Purchased"]
df["Updated_Purchased"].fillna("No", inplace = True)

# Other Ways can be to fill the missing values using the following methods

# Assigning a Constant Value , this is for Simple Imputation

#example

df["Updated_Age_With_Constant_Value"] = df["Age"]
df["Updated_Age_With_Constant_Value"].fillna(10, inplace = True)

print(df)
```

```python
# Simple Imputation Example Using Scikit Learn Library

from sklearn.impute import SimpleImputer
import numpy as np

# Loading Data again

data = pd.read_csv('Data.csv')
dataFrame = pd.DataFrame(data)
# print(dataFrame)

Age_imputer = SimpleImputer(missing_values=pd.NA,strategy='mean')
Salary_Imputer = SimpleImputer(missing_values=pd.NA,strategy='median')
Country_Imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')

dataFrame["Age_Imputed_Values"] = Age_imputer.fit_transform(dataFrame[["Age"]])
dataFrame["Salary_Imputed_Values"] = Salary_Imputer.fit_transform(dataFrame[["Salary"]])
dataFrame['Country_Imputed_Values'] = Country_Imputer.fit_transform(dataFrame[['Country']]).ravel()
print(dataFrame)


```

```python
#using KNN Imputer

import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Create an instance of KNNImputer
knn_imputer = KNNImputer(n_neighbors=2, weights="uniform", missing_values=np.nan)

dataFrame["Salary_Imputed_KNN"] = knn_imputer.fit_transform(dataFrame[["Salary"]])
dataFrame["Age_Imputed_KNN"] = knn_imputer.fit_transform(dataFrame[["Age"]])

print(dataFrame)
```

```python
# Checking for Normal Distribution
# Data is said to be normally Distributed if Frequency Distribution of Data has Bell Curve
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, anderson
from scipy import stats

# Histogram for Age and Salary
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(dataFrame['Age_Imputed_Values'].dropna(), kde=True)
plt.title('Age Distribution')

plt.subplot(1, 2, 2)
sns.histplot(dataFrame['Salary_Imputed_Values'].dropna(), kde=True)
plt.title('Salary Distribution')

plt.show()

# Q-Q Plot for Age and Salary
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
stats.probplot(dataFrame['Age_Imputed_Values'].dropna(), dist="norm", plot=plt)
plt.title('Q-Q Plot for Age')

plt.subplot(1, 2, 2)
stats.probplot(dataFrame['Salary_Imputed_Values'].dropna(), dist="norm", plot=plt)
plt.title('Q-Q Plot for Salary')

plt.show()
```

```python
# Shapiro-Wilk Test
age_shapiro_test = shapiro(dataFrame['Age_Imputed_Values'].dropna())
salary_shapiro_test = shapiro(dataFrame['Salary_Imputed_Values'].dropna())

print(f'Age Shapiro-Wilk Test: Statistic={age_shapiro_test.statistic}, p-value={age_shapiro_test.pvalue}')
print(f'Salary Shapiro-Wilk Test: Statistic={salary_shapiro_test.statistic}, p-value={salary_shapiro_test.pvalue}')

# Anderson-Darling Test
age_anderson_test = anderson(dataFrame['Age_Imputed_Values'].dropna())
salary_anderson_test = anderson(dataFrame['Salary_Imputed_Values'].dropna())

print(f'Age Anderson-Darling Test: Statistic={age_anderson_test.statistic}, Critical Values={age_anderson_test.critical_values}, Significance Levels={age_anderson_test.significance_level}')
print(f'Salary Anderson-Darling Test: Statistic={salary_anderson_test.statistic}, Critical Values={salary_anderson_test.critical_values}, Significance Levels={salary_anderson_test.significance_level}')

stats.anderson(dataFrame['Age_Imputed_Values'].dropna(), dist='logistic')
```

```python
#Comparative Tests

from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns

# Create a contingency table
contingency_table = pd.crosstab(dataFrame['Country_Imputed_Values'], dataFrame['Purchased'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-square Statistic: {chi2}, p-value: {p}')

# Visualize the contingency table
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Contingency Table: Country vs Purchased')
plt.show()
```
